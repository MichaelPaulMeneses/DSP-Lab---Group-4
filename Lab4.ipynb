```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    y = np.exp(-t/2)
    return y
```

```python
N = np.linspace(-10, 10, 21) ##Code 6

def impulse(N):
    return np.where(N == 0, 1, 0)

plt.stem(N, impulse(N), linefmt='Orange', label='Impulse Stem')
plt.plot(N, impulse(N), color='Black', label='Impulse Plot')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
def ramp(A,B,C): ##Code 7
    return B*A+C

plt.plot(N, ramp(N, 1, 0), label="m=1, b=0")
plt.plot(N, ramp(N, 0.5, 0), label="m=0.5, b=0")
plt.plot(N, ramp(N, 1, 2), label="m=1, b=2")
plt.plot(N, ramp(N, -0.5, 0), label="m=-0.5, b=0")

plt.axhline(color="k")
plt.axvline(color="k")

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
N = np.linspace(-2, 2, 21) ##Code 8

plt.figure(figsize=(10, 6))
plt.stem(N, np.exp(N), linefmt='orange', label=r'$e^n$')
plt.stem(N, np.exp(-N), linefmt='blue', label=r'$e^{-n}$')
plt.stem(N, -np.exp(N), linefmt='orangered', label=r'$-e^n$')
plt.stem(N, -np.exp(-N), linefmt='dodgerblue', label=r'$-e^{-n}$')

plt.legend(loc='lower center', title='Stem Plots')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
N = np.linspace(-10, 10, 21) ##Code 9

def step(N):
    return np.where(N >=0, 1, 0)

plt.stem(N, step(N,), label='Step Stem')
plt.plot(N, step(N), color='Orange', label='Step Plot')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
T = np.linspace(-2*np.pi, 2*np.pi, 100) ##Code 10
T2 = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.figure(figsize=(10, 4))
plt.title('Amplitude Shift')

plt.plot(T, np.sin(T), label=r'$\sin \theta$', color='Yellow')
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='Purple')

plt.plot(T2, 2*np.sin(T2), label=r'$2\sin \theta$') 
plt.stem(T2, 2*np.sin(T2), label=r'$2\sin \theta$', linefmt='Green')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 4)) ##Code 11
plt.title('Frequency Shift')
plt.plot(T, np.sin(T), label=r'$\sin \theta$', color="Blue")
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='Green')

plt.plot(T, np.sin(2*T), label=r'$\sin 2\theta$', color='Orange') 
plt.stem(T2, np.sin(2*T2), label=r'$\sin 2\theta$', linefmt='Purple')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 4)) ##Code 12
plt.title('Phase Shift')
plt.plot(T, np.sin(T), label=r'$\sin \theta$', color='green')
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='Orangered')

plt.plot(T2, np.sin(T2+np.pi/4), label=r'$\sin \theta + \frac{\pi}{4}$')
plt.stem(T2, np.sin(T2+np.pi/4), label=r'$\sin \theta + \frac{\pi}{4}$', linefmt='Black')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
def euler(theta):
    return np.exp(1j*theta)
```

```python
T = np.linspace(-2*np.pi, 2*np.pi, 100) ## Code 13

plt.figure(figsize=(10, 4))
plt.plot(T, euler(T).real, label='Real')
plt.plot(T, euler(T).imag, label='Imaginary', color='orange')

plt.stem(T, euler(T).real, label='Real', markerfmt='ro', basefmt='r')
plt.stem(T, euler(T).imag, label='Imaginary', markerfmt='bo', basefmt='b')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
