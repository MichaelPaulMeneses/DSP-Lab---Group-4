```python
n = np.linspace(0, 99, 100) ##Code 2
x_n = np.sin(0.2*np.pi*n)
y_n = x_n**2

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.scatter(n, x_n, color='Orange', label='Input Signal x[n]')
plt.title("Input Signal x[n] = sin(0.2*pi*n)")
plt.xlabel('Sample Number [n]')
plt.ylabel('Amplitude')
plt.grid()

plt.subplot(2, 1, 2)
plt.bar(n, y_n, color='Blue', label='Output Signal y[n]')
plt.title("Output Signal y[n]=x[n]**2")
plt.xlabel('Sample Number [n]')
plt.ylabel('Amplitude')
plt.grid()
plt.tight_layout()
plt.legend()

plt.figure(figsize=(10, 6))
plt.plot(n, x_n, color='Orange', label='Input Signal x[n]')
plt.plot(n, y_n, color='Blue', label='Output Signal y[n]')
plt.title('Input and Output Signals')
plt.xlabel('Sample Number [n]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.show()
```

```python
n_samples = 100 ##Code 3
n = np.arange(n_samples)
x_clean = np.sin(0.3 * np.pi * n)
noise = np.random.normal(0, 0.5, n_samples)
x_noisy = x_clean + noise
y = np.zeros(n_samples)
for i in range(1, n_samples):
    y[i] = 0.5 * y[i-1] + 0.5 * x_noisy[i]

plt.figure(figsize=(10, 6))
plt.subplot(3,1,1)
plt.plot(n, x_clean, label='Clean Input', color='violet')
plt.title('Clean Input')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3,1,2)
plt.plot(n, x_noisy, label='Noisy Input Signal', color='brown')
plt.title('Noisy Input Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(n, y, label='System Output', color='black')
plt.title('System Output')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
