```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000) ##Code 1
x_t = np.sin(t)
y_t = 2 * x_t
tau = 2

t_shifted = t - tau
x_t_shift = np.sin(t -2)
y_t_shift = 2 * x_t_shift
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, y_t, label= 'y(t) = 2 * x(t)', color="orange")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()

plt.plot(t, y_t_shift, label="y(t - ) = 2 * x(t - )", color="blue")
plt.title("Original and Time-Shifted Signal (Multiply by a Constant)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) ##Code 2
x_t = np.sin(t)
y_t = t * x_t
tau = 2

t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = (t_shifted)*x_t_shift
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, y_t, label= 'y(t) = t * x(t)', color="orange")
plt.xlabel("X")
plt.grid()


plt.plot(t, y_t_shift, label="y(t - ) = (t - ) * x(t - )", color="blue")
plt.title("Original and Time-Shifted Signal (Multiply by Time)")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) ##Code 3
x_t = np.sin(t)
y_t = np.gradient(x_t, t)
tau = 2

t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = np.gradient(x_t_shift, t)
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, y_t, label="y(t) = Velocity * x(t)", color='orange')
plt.ylabel("Y")

plt.plot(t, y_t_shift, label="y(t - ) = Velocity * x(t - )", color='blue')
plt.title("Original and Time-Shifted Signal (Differentiation)")
plt.xlabel("X")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
