```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100 ##Code 1
x = np.linspace(0, 99, n)
signal = np.sin(2 * np.pi * x / 100)
noise = np.random.normal(0, 1, n)
input_signal = signal + noise
y = np.zeros(n)
for i in range(2, n):
    y[i] = (1/3) * (input_signal[i] + input_signal[i-1] + input_signal[i-2])

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x, input_signal, label='Noisy Input Signal')
plt.xlabel('Sample n')
plt.ylabel('Amplitude')
plt.title('Digital Filtering in Signal Processing (Causal System)')
plt.legend()

plt.plot(x, y, label='Filtered Signal')
plt.xlabel('Sample n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
n = np.linspace(0, 99, 100) ##Code 2
base_price = 100
stock_prices = base_price + np.random.normal(0, 5, 100)

y = np.zeros(100)
for i in range(1, 99):
    y[i] = (1/3) * (stock_prices[i-1] + stock_prices[i] + stock_prices[i+1])

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(n, stock_prices, label='Original Stock Prices')
plt.xlabel('Sample n')
plt.ylabel('Price')
plt.title('Stock Price Prediction (Non-Causal System)')
plt.legend()

plt.plot(n, y, label='Smoothed Stock Prices')
plt.xlabel('Sample n')
plt.ylabel('Price')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x = np.arange(100) ##Code 3
signal = np.sin(2 * np.pi * x / 100)
echo_delay = 5
echo_attenuation = 0.6
echo_signal = np.zeros_like(signal)
echo_signal[echo_delay:] = signal[:-echo_delay] * echo_attenuation
input_signal = signal + echo_signal


y = input_signal
y[echo_delay:] -= signal[:-echo_delay] * echo_attenuation

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(x, input_signal, label='Input Signal with Echo', color='blue')
plt.xlabel('Sample n')
plt.ylabel('Amplitude')
plt.title('Echo Cancellation in Telecommunications (Causal System)')
plt.legend()

plt.plot(x, y, label='Echo-Canceled Signal', color='orange')
plt.plot(x, signal, label='Original Signal',color='green')
plt.xlabel('Sample n')
plt.ylabel('Amplitude')
plt.grid()
plt.tight_layout()
plt.legend()
plt.show()
```
