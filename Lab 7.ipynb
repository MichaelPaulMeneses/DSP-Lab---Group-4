```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(0, 100)
clean_signal = np.sin(2 * np.pi * 0.05 * n)
noise = np.random.normal(0, 0.5, 100)
noisy_signal = clean_signal + noise
filtered_signal = np.zeros_like(noisy_signal)

for i in range(2, len(noisy_signal)):
    filtered_signal[i] = (noisy_signal[i] + noisy_signal[i-1] + noisy_signal[i-2]) / 3

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n,noisy_signal, label='Noisy input signal')
plt.plot(n,filtered_signal, label='Filtered signal')
plt.xlabel('Sample index')
plt.ylabel('Signal amplitude')
plt.title('Digital Filtering in Signal Processing')
plt.legend()
plt.grid()
plt.show()
```

```python
n = np.arange(100)
base_price = 100  
stock_prices = base_price + np.random.normal(0, 2, len(n))
smoothed_prices = np.zeros_like(stock_prices)

for i in range(1, len(n)-1): 
    smoothed_prices[i] = (stock_prices[i-1] + stock_prices[i] + stock_prices[i+1]) / 3
smoothed_prices[0] = stock_prices[0]
smoothed_prices[-1] = stock_prices[-1]

plt.figure(figsize=(10, 6))
plt.plot(stock_prices, label='Original Stock Prices', color='orange')
plt.plot(smoothed_prices, label='Smoothed Stock Prices', color='blue')
plt.title('Stock Price Prediction with Non-Causal Moving Average Filter')
plt.xlabel('Day')
plt.ylabel('Stock Price')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
n_samples = 100  ##Code 3
n = np.arange(n_samples)
x = np.sin(0.1 * np.pi *  n)
d = 5
alpha = 0.6  
echo = np.zeros(n_samples)
echo[d:] = alpha * x[:-d]  
x_with_echo = x + echo
y = np.copy(x_with_echo) 
y[d:] -= alpha * x[:-d]

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(n, x_with_echo, label='Signal with Echo', color='orange')
plt.title("Echo Cancellation in Telecommunications(Casual System)")
plt.ylabel('Amplitude')
plt.legend()

plt.plot(n, y, label='Echo-cancelled Signal', color='blue')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
