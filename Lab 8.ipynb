```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
```

```python
t = np.linspace(-1, 1, 500)
f0 = 5  # Frequency of the signal

signal = np.sin(2 * np.pi * f0 * t)

signal_fft = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), d=(t[1] - t[0]))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(1, 2, 2)
plt.plot(frequencies, np.abs(signal_fft))
plt.title("Frequency Domain (Magnitude)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

```python
f1, f2 = 5, 10
composite_signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

composite_fft = np.fft.fft(composite_signal)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, composite_signal)
plt.title("Time Domain Composite Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Frequency Domain
plt.subplot(1, 2, 2)
plt.plot(frequencies, np.abs(composite_fft))
plt.title("Frequency Domain of Composite Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

```python
sample_rate, data = read('file_example_WAV_1MG.wav')
if len(data.shape) > 1:
    data = data.mean(axis=1)
time = np.arange(len(data)) / sample_rate

t = time
f = np.fft.fftfreq(len(data), d=1/sample_rate)
data_fft = np.fft.fft(data)
magnitude_spectrum = np.abs(data_fft)

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(time, data)
plt.title("Time Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(f, magnitude_spectrum, color='orange')
plt.title("Frequency Domain (Magnitude)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.io.wavfile import write

# Define parameters
sample_rate = 44100  # samples per second
duration = 2  # seconds
frequency = 440  # A4 note in Hz (440 Hz)

# Generate time values
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
# Generate sine wave signal
signal = 0.5 * np.sin(2 * np.pi * frequency * t)

# Save the signal as a .wav file
write("sample.wav", sample_rate, signal.astype(np.float32))
```
