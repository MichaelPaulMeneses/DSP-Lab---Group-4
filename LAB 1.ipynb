```python
class Student:
    def __init__(self, Student, Course, Prelim, Midterm, Final):
        self.Student = Student
        self.Course = Course
        self.Prelim = Prelim
        self.Midterm = Midterm
        self.Final = Final

    def myFunction(self):
        print("\nName: ", self.Student, "\n\nCourse: ", self.Course, "\n")

    def calculate_Semestral_grade(self):
        return (0.3 * self.Prelim + 0.3 * self.Midterm + 0.4 * self.Final)

    def get_grade_equivalent(self, grade):
        if 97 <= grade <= 100:
            return "1.00"
        elif 93 <= grade < 97:
            return "1.25"
        elif 89 <= grade < 93:
            return "1.50"
        elif 85 <= grade < 89:
            return "1.75"
        elif 82 <= grade < 85:
            return "2.00"
        elif 79 <= grade < 82:
            return "2.25"
        elif 76 <= grade < 79:
            return "2.50"
        elif 73 <= grade < 76:
            return "2.75"
        elif 70 <= grade < 73:
            return "3.00"
        else:
            return "Failed"

Student_Name = input("Enter your name: ")
Course_Name = input("Enter your course name: ")

Prelim_grade = float(input("\n" "Enter your Prelim Grade: "))
Midterm_grade = float(input("\n" "Enter your Midterm Grade: "))
Final_grade = float(input("\n" "Enter your Final Grade: "))


myself = Student(Student_Name, Course_Name, Prelim_grade, Midterm_grade, Final_grade)
myself.myFunction()

Semestral_grade = myself.calculate_Semestral_grade()
print("Semestral Grade: {:.2f}".format(Semestral_grade))


happy, lol, sad = "\U0001F600", "\U0001F606", "\U0001F62D"

if Semestral_grade > 70.00:
    print(f"Result: {happy}" "\n")

elif Semestral_grade == 70.00:
    print(f"Result: {lol}" "\n")

else:
    print(f"Result: {sad}" "\n")


show_equivalent = input("Do you want to see your grade equivalent? (yes/no): ")
if show_equivalent == "yes":
    grade_equivalent = myself.get_grade_equivalent(Semestral_grade)
    print(f"Your grade equivalent is: {grade_equivalent}")
```
