```python
import numpy as np
from scipy import signal
from scipy.fft import fft, fftfreq, fft2, fftshift
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 500) ##1
x = np.sin(2 * np.pi * 5 * t) 
averaging_filter = np.ones(10) / 10
y = signal.convolve(x, averaging_filter, mode='same')
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Original Signal (Sine Wave)')
plt.plot(t, y, label='After Convolution (Averaging Filter)')
plt.legend()
plt.grid()
plt.title('1D Convolution with Averaging Filter')
plt.show()
```

```python
matrix = np.random.random((10, 10)) ##2
box_filter = np.ones((3, 3)) / 9

convolved_matrix = signal.convolve2d(matrix, box_filter, mode='same')
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(matrix, cmap='inferno')
axes[0].set_title('Original Matrix')
axes[1].imshow(convolved_matrix, cmap='inferno')
axes[1].set_title('After Convolution (Box Filter)')
plt.show()
```

```python
x = np.sin(2 * np.pi * 7 * t) + 0.5 * np.random.randn(len(t)) ##3
X = np.fft.fft(x)
frequencies = np.fft.fftfreq(len(t), d=t[1] - t[0])

plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X))
plt.title('Frequency Response (FFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(np.abs(X), bins=30, edgecolor='black')
plt.title('Histogram of Frequency Response')
plt.xlabel('Amplitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
matrix = np.random.random((10, 10)) ##4
fft_matrix = np.fft.fft2(matrix)
fft_matrix_shifted = np.fft.fftshift(fft_matrix)
magnitude_spectrum = np.abs(fft_matrix_shifted)

plt.figure(figsize=(6, 4))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='RdBu')
plt.title('2D Magnitude Spectrum (Log Scale)')
plt.grid()
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(magnitude_spectrum.ravel(), bins=50, edgecolor='black')
plt.title('Histogram of 2D Magnitude Spectrum')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
t = np.linspace(0, 1, 300) ##5
x = np.sin(2 * np.pi * 7 * t) + 0.6* np.random.randn(len(t))
moving_avg_filter = np.ones(12) / 10

y = signal.convolve(x, moving_avg_filter, mode='same')
plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Noisy Signal')
plt.plot(t, y, label='Smoothed Signal (Moving Avg)')
plt.legend()
plt.title('1D Convolution with Moving Average Filter')
plt.grid()
plt.show()
```

```python
matrix = np.random.random((12, 12)) ##6
box_filter = np.ones((3, 3)) / 12

convolved_matrix = signal.convolve2d(matrix, box_filter, mode='same')
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(matrix, cmap='RdBu')
axes[0].set_title('Original Matrix')
axes[1].imshow(convolved_matrix, cmap='RdBu')
axes[1].set_title('After Convolution (Box Filter)')
plt.show()
```

```python
x = np.sin(2 * np.pi * 3 * t) ##7
derivative_filter = np.array([1, 0, -1])
y = signal.convolve(x, derivative_filter, mode='same')

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original Signal')
plt.plot(t, y, label='Derivative')
plt.legend()
plt.title('1D Convolution with Derivative Filter')
plt.grid()
plt.show()
```

```python
x = np.sin(2 * np.pi * 10 * t) + 0.5 * np.random.randn(len(t)) ##8
X = np.fft.fft(x)

frequencies = np.fft.fftfreq(len(t), d=t[1] - t[0])
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X), color='orange')
plt.title('Frequency Response (FFT)')
plt.xlabel('Frequency (Hz)')
plt.grid()
plt.ylabel('Amplitude')
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(np.abs(X), bins=30, edgecolor='black')
plt.title('Histogram of Frequency Response')
plt.xlabel('Amplitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
matrix = np.random.random((12, 12)) ##9
fft_matrix = np.fft.fft2(matrix)
fft_matrix_shifted = np.fft.fftshift(fft_matrix)
magnitude_spectrum = np.abs(fft_matrix_shifted)

plt.figure(figsize=(6, 4))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='RdBu')
plt.title('2D Magnitude Spectrum (Log Scale)')
plt.colorbar()
plt.grid()
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(magnitude_spectrum.ravel(), bins=50, edgecolor='black', color='pink')
plt.title('Histogram of 2D Magnitude Spectrum')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
x = np.concatenate([np.zeros(50), np.ones(50)]) ##10
edge_filter = np.array([-1, 1])
y = signal.convolve(x, edge_filter, mode='same')

plt.figure(figsize=(10, 4))
plt.plot(x, label='Original Step Signal')
plt.plot(y, label='After Edge Detection')
plt.legend()
plt.title('1D Convolution for Edge Detection')
plt.grid()
plt.show()
```
