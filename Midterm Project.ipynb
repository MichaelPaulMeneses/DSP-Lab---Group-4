```python
import numpy as np
from scipy import signal #To run the SciPy Convolution
from scipy.signal import convolve as scipy_convolve # To run the Scipy Convolve that differentiate numpy and scipy
from scipy.fft import fft, fftfreq, fft2, fftshift #To run the scipy fft(Fast Fourier Transform)
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 500) # Problem Set 1
x = np.sin(2 * np.pi * 5 * t) # 5 Hz wave

# Define a simple averaging filter (length = 20)
averaging_filter = np.ones(20) / 20

# Perform 1D convolution using scipy.signal.convolve
y = signal.convolve(x, averaging_filter, mode='same')

#Plot the Original and Convolved Signals
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Original Signal (Sine Wave)')
plt.plot(t, y, label='After Convolution (Averaging Filter)')
plt.legend()
plt.grid()
plt.title('1D Convolution with Averaging Filter')
plt.show()
```

```python
#10x10 Random Matrix (grayscale image)
matrix = np.random.random((10, 10)) #Problem Set 2

#3x3 Box Filter
box_filter = np.ones((3, 3)) / 9

#Perform 2D convolution using scipy.signal.convolve2d
convolved_matrix = signal.convolve2d(matrix, box_filter, mode='same')

# Plot the Figures
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(matrix, cmap='gray')
axes[0].set_title('Original Matrix')
axes[1].imshow(convolved_matrix, cmap='gray')
axes[1].set_title('After Convolution (Box Filter)')
plt.show()
```

```python
#7 Hz wave
x = np.sin(2 * np.pi * 7 * t) + 0.5 * np.random.randn(len(t)) #Problem Set 3

#Perform the FFT(Fast Fourier Transform)
X = np.fft.fft(x)

#Freequency Bins
frequencies = np.fft.fftfreq(len(t), d=t[1] - t[0])

#Plot the Frequency
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(X))
plt.title('Frequency Response (FFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

#Plot the Histogram
plt.figure(figsize=(6, 4))
plt.hist(np.abs(X), bins=30, edgecolor='black')
plt.title('Histogram of Frequency Response')
plt.xlabel('Amplitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
#10x10 Random Matrix
matrix = np.random.random((10, 10)) #Problem Set 4

# Perform 2D FFT on the matrix
fft_matrix = np.fft.fft2(matrix)
fft_matrix_shifted = np.fft.fftshift(fft_matrix)

#Magnitude Spectrum
magnitude_spectrum = np.abs(fft_matrix_shifted)

#Plot the 2D Magnitude Spectrum
plt.figure(figsize=(6, 4))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='RdBu')
plt.title('2D Magnitude Spectrum (Log Scale)')
plt.grid()
plt.show()

#Plot the Histogram
plt.figure(figsize=(6, 4))
plt.hist(magnitude_spectrum.ravel(), bins=50, edgecolor='black')
plt.title('Histogram of 2D Magnitude Spectrum')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
# Generate a noisy sine wave
t = np.linspace(0, 1, 300) #Problem Set 5
x = np.sin(2 * np.pi * 10 * t) + 0.6* np.random.randn(len(t))

# Moving average filter of length 12
moving_avg_filter = np.ones(12) / 10

# Perform 1D convolution
y = signal.convolve(x, moving_avg_filter, mode='same')

#Plot the Noisy and Smoothed Signals
plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Noisy Signal')
plt.plot(t, y, label='Smoothed Signal (Moving Avg)')
plt.legend()
plt.title('1D Convolution with Moving Average Filter')
plt.grid()
plt.show()
```

```python
# 10x10 matrix
matrix = np.random.random((12, 12)) #Problem Set 6

#6x6 Box Filter
box_filter = np.ones((6, 6)) / 12

# Perform 2D convolution using scipy.signal.convolve2d
convolved_matrix = signal.convolve2d(matrix, box_filter, mode='same')

#Plot the Figures 
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(matrix, cmap='RdBu')
axes[0].set_title('Original Matrix')
axes[1].imshow(convolved_matrix, cmap='RdBu')
axes[1].set_title('After Convolution (Box Filter)')
plt.show()
```

```python
#Generate Sine Wave
x = np.sin(2 * np.pi * 3 * t) #Problem Set 7

#Derivative Filter
derivative_filter = np.array([1, 0, -1])

# 1D SciPy Convolution to get the derivative
y = signal.convolve(x, derivative_filter, mode='same')

#Plot the Figures
plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original Signal')
plt.plot(t, y, label='Derivative')
plt.legend()
plt.title('1D Convolution with Derivative Filter')
plt.grid()
plt.show()
```

```python
# 12x12 Random Matrix
matrix = np.random.random((12, 12)) #Problem Set 9

#Applying the matrix
fft_matrix = np.fft.fft2(matrix)
fft_matrix_shifted = np.fft.fftshift(fft_matrix)
magnitude_spectrum = np.abs(fft_matrix_shifted)


# Plot the Figure
plt.figure(figsize=(6, 4))
plt.imshow(np.log(1 + magnitude_spectrum), cmap='RdBu')
plt.title('2D Magnitude Spectrum (Log Scale)')
plt.colorbar()
plt.grid()
plt.show()

# Plot the Figure
plt.figure(figsize=(6, 4))
plt.hist(magnitude_spectrum.ravel(), bins=50, edgecolor='black', color='brown')
plt.title('Histogram of 2D Magnitude Spectrum')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.grid()
plt.show()
```
